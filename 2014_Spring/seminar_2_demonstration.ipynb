# Генерация данных

```python
import numpy as np
import pylab as pl
```

Сгенерируем данные из двух двумерных нормальных распределений и визуализируем их

```python
n = 100
a = np.random.normal(loc=0, scale=1, size=(n, 2))
b = np.random.normal(loc=3, scale=2, size=(n, 2))
pl.scatter(a[:, 0], a[:, 1], c='blue', s=50, alpha=0.6)
pl.scatter(b[:, 0], b[:, 1], c='red', s=50, alpha=0.6)
```

Объединим матрицы в единую матрицу данных и создадим отдельный столбец меток.

```python
X = np.vstack([a, b])
print X.shape
y = np.hstack([np.zeros(n), np.ones(n)])
print y.shape
```

Можно визуализировать одной командой scatter

```python
pl.scatter(X[:, 0], X[:, 1], c=y, s=50, alpha=0.6)
```

# Обучение модели

Обучим метод 3ех ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
clf = KNN(n_neighbors=3)
clf.fit(X, y)
print clf
```

Предскажем метки для каких-нибдуь точек.

```python
clf.predict_proba([
                   [1, 1],
                   [10, 20],
                   [-10, -20]
                   ])
```

Напишем функцию для генерации карты точек.

```python
def get_grid(data, border=1., step=.01):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step),
                         np.arange(y_min, y_max, step))
```

Обучим классификатор на карте и визуализируем ее

```python
xx, yy = get_grid(X)
predicted = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)
pl.pcolormesh(xx, yy, predicted)
pl.scatter(X[:, 0], X[:, 1], c=y, s=100)
```

Немного побалуемся с параметрами визуализации

```python
pl.figure(figsize=(10, 7))
pl.pcolormesh(xx, yy, predicted, cmap='spring')
pl.scatter(X[:, 0], X[:, 1], c=y, s=150, cmap='spring', alpha=0.7)
```

Оберенем все вышеописанное в одну функцию

```python
def plot_model(X, y, clf):
    clf.fit(X, y)
    xx, yy = get_grid(X)
    predicted = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)
    pl.figure(figsize=(10, 7))
    pl.pcolormesh(xx, yy, predicted, cmap='spring')
    pl.scatter(X[:, 0], X[:, 1], c=y, s=150, cmap='spring', alpha=0.7)
```

Визуализируем работу 1-NN

```python
plot_model(X, y, KNN(n_neighbors=1))
```

Визуализируем работу 15-NN

```python
plot_model(X, y, KNN(n_neighbors=15))
```

Сколько ближайших соседей надо брать? Нужно сравнивать качество!

# Оценка качества

Разобьем данные на обучение и контроль

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, clf.predict(X_test))
```

Оценим качество для каждого из k.

```python
scores = []
for k in range(1, 51):
    scores.append(accuracy_score(y_test, KNN(n_neighbors=k).fit(X_train, y_train).predict(X_test)))
pl.plot(scores)
```

Не ясно, какое значение параметра оптимально. Проведем кроссвалидацию уже готовыми библиотечными средствами

```python
from sklearn.grid_search import GridSearchCV
params = {'n_neighbors': range(1, 51)}
grid_searcher = GridSearchCV(clf, params, cv=5)
grid_searcher.fit(X, y)
```

Визуализируем полученные оценки качества и их $1\sigma$-доверительный интервал

```python
means = []
stds = []
for elem in grid_searcher.grid_scores_:
    means.append(np.mean(elem.cv_validation_scores))
    stds.append(sqrt(np.var(elem.cv_validation_scores)))
means = np.array(means)
stds = np.array(stds)
```

```python
pl.figure(figsize=(8, 6))
pl.plot(means)
pl.fill_between(range(len(means)), means + stds, means - stds, alpha = 0.1, facecolor='blue')
pl.xlabel('Number of nearest neighbours')
pl.ylabel('Accuracy')
```

Здесь уже лучше видно, какое значение параметра лучше относительно заданной метрики качества. Стоит брать 3-4 ближайших соседа.
