## Часть вторая. Baseline

Попробуем воспроизвести бейзлайн, т.е. в качестве предсказания возвращать последнее наблюдаемое значение. Чтобы пощупать устойчивость этого решения, проведем 100 произвольных разбиений на обучение и тест и оценим качество на этих разбиениях.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.cross_validation import ShuffleSplit

def score_baseline_model(x, y):
    clfs = []
    scores = []    
    ids = ShuffleSplit(x.shape[0], n_iter=100, test_size=0.25)
    for train_id, test_id in ids:
        train_x, test_x = x[train_id, :], x[test_id, :]
        train_y, test_y = y[train_id, :], y[test_id, :]
            
        test_y_predicted = test_x[:, -244-198:-244].ravel() # last obeserved value
        score = mean_absolute_error(test_y.ravel(), test_y_predicted.ravel())
        scores.append(score)
    return scores
      
baseline_scores = score_baseline_model(train_x, train_y)
print baseline_scores
```

Посмотрим, как выглядят визуально эти значения, чему равно среднее и дисперсия.

```python
pl.plot(baseline_scores)
pl.plot(np.mean(baseline_scores) * np.ones(len(baseline_scores)), c='red')
print np.mean(baseline_scores)
print np.var(baseline_scores)
```

На лидерборде качество бэйзлайна равно 0.42596, что вполне соответствует ожиданиям.

## Часть третья. Ridge Regression, или Все Плохо

Для начала попробуем воспользоваться линейной моделью, а именно - ридж-регрессией. В качестве оптимизируемого параметра выберем степень регуляризации alpha. Для каждого из предсказываемых выходов обучим независимую ридж-регрессию на данных.

```python
from sklearn.linear_model import Ridge
from collections import defaultdict


def score_ridge_model(x, y):
    clfs = []
    scores_ = []    
    
    for i in range(y.shape[1]):
        print i,
        y_ = y[:, i].ravel()
        alphas = [0.1, 1.0, 10.0, 100.0, 1000.0]
        scores = defaultdict(list)
        
        ids = ShuffleSplit(x.shape[0], n_iter=3, test_size=0.25)
        
        for train_id, test_id in ids:
            train_x, test_x = x[train_id, :], x[test_id, :]
            train_y, test_y = y_[train_id, :], y_[test_id, :]

            for alpha in alphas:
                clf = Ridge(alpha=alpha)
                clf.fit(train_x, train_y)
                score = mean_absolute_error(test_y, clf.predict(test_x))
                scores[alpha].append(score)
        scores_.append(scores)  
    return scores_
      
ridge_scores = score_ridge_model(train_x, train_y)
```

Посчитаем качество линейной модели. Оно фиговое. Вряд ли она нам подойдет.

```python
res_ridge_scores = []
for score in ridge_scores:
    means = []
    for alpha in score:
        means.append(np.mean(score[alpha]))
    res_ridge_scores.append(min(means))
print np.mean(res_ridge_scores)
```

Но коль уж у нас есть значения качества посчитанные для разных значений alpha, то можно посмотреть, какое alpha нас больше всего устраивает. Оказывается, большое alpha в среднем самое хорошее. Наверное, можно было бы дальше подбирать оптимальное значение alpha, но я этим не занимался (хотя бы потому что для каждого признака alpha может быть свое).

```python
scores_alpha = defaultdict(list)
for score in ridge_scores:
    for alpha in score:
        scores_alpha[alpha].append(np.mean(score[alpha]) / np.mean(score[1.0]))
pl.figure(figsize=(15, 5))

plt.subplot2grid((1, 2), (0, 0))
pl.plot(scores_alpha[0.1], c='blue')
pl.plot(scores_alpha[1], c='red')
pl.plot(scores_alpha[10], c='green')
pl.plot(scores_alpha[100], c='purple')
#pl.plot(scores_alpha[1000], c='black')

plt.subplot2grid((1, 2), (0, 1))
pl.plot(scores_alpha[0.1], c='blue')
pl.plot(scores_alpha[1], c='red')
pl.plot(scores_alpha[10], c='green')
pl.plot(scores_alpha[100], c='purple')
pl.plot(scores_alpha[1000], c='black')
```

Интересная особенность. График качества по всем предсказываемым признакам выглядит очень похоже (с точностью до растяжения) при разных значениях параметра регуляризации alpha.

## Часть последняя. Попытки добиться хоть каких-то результатов

Последняя попытка добиться каких-то результатов. Давайте для каждой предсказываемой метки выбирать оптимум между линейной моделью, случайным лесом и последним наблюдаемым значением. Авось для какой-нибудь метки удастся найти более удачный метод, нежели бейзлайн.

```python
def score_common_model(x, x_pca, y):
    scores_ = []    
    
    for i in range(y.shape[1]):
        print i, 
        y_ = y[:, i].ravel()
        scores = defaultdict(list)
        
        ids = ShuffleSplit(x.shape[0], n_iter=5)
        
        for train_id, test_id in ids:
            train_x, test_x = x[train_id, :], x[test_id, :]
            train_x_pca, test_x_pca = x_pca[train_id, :], x_pca[test_id, :]
            train_y, test_y = y_[train_id, :], y_[test_id, :]
            
            clf = RandomForestRegressor(n_estimators=200)
            clf.fit(train_x_pca, train_y)    
            score = mean_absolute_error(test_y, clf.predict(test_x_pca))
            scores['rf'].append(score)
            
            clf = Ridge(alpha=1000)
            clf.fit(train_x, train_y)    
            score = mean_absolute_error(test_y, clf.predict(test_x))
            scores['ridge'].append(score)
            
            test_y_predicted = test_x[:, -244-198+i].ravel()
            score = mean_absolute_error(test_y, test_y_predicted)
            scores['baseline'].append(score)
        scores_.append(scores)
    return scores_
common_scores = score_common_model(train_x, train_x_pca, train_y)
```

Посмотрим, есть ли метки, на которых качество было выше бейзлайна.

```python
res_scores = []
for score in common_scores:
    means = []
    for alpha in score:
        means.append(np.mean(score[alpha]))
    res_scores.append(min(means))
    if min(means) != np.mean(score['baseline']):
        print 'Yahoo!!!'
print np.mean(res_scores)
```
